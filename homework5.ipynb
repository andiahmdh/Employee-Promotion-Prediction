```python
df['Income_Category'].value_counts()
```

```python
#data attrition_counts
attrition_counts = df['Attrition_Flag'].value_counts()
my_colors1 = ["#27aeef", "#b33dc6"]

#data customer_age_counts
customer_age_counts = df['Customer_Age'].value_counts()
my_colors2 = ["#ea5545", "#f46a9b", "#ef9b20"]

#data customers_gender
customers_gender = df['Gender'].value_counts()

#data education_level
education_level = df['Education_Level'].value_counts()

#data income_category
income_category = df['Income_Category'].value_counts()


### Visualization Proccess ###

#text style
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 15,
        }

fig = plt.figure(figsize=(22,10) )

fig.suptitle('Data Visualization', fontdict=font)

ax1 = fig.add_subplot(231)
ax1.set_title('Attrition Flag', fontdict=font)
ax1.pie(attrition_counts.values,
        labels=attrition_counts.index,
        autopct='%1.1f%%',
        colors=my_colors1,
        shadow=True,
        startangle=130,
        wedgeprops= {'edgecolor':'white', 
                     'linewidth': 3, 
                     'antialiased': True},
        textprops= {'family':'serif',
                    'color':'black',
                    'weight':'normal', 
                    'fontsize':10})


ax2 = fig.add_subplot(232)
ax2.set_title('Customer Age', fontdict=font)
ax2.set_xlabel('Age', fontdict=font, fontsize=11)
ax2.set_ylabel('Numbers', fontdict=font, fontsize=11)
ax2.grid(False)
ax2.hist(df['Customer_Age'],
         alpha=0.9,
         facecolor='#bdcf32')

   
ax3 = fig.add_subplot(233)
ax3.set_title('Customer Gender', fontdict=font)
ax3.pie(customers_gender.values,
        labels=customers_gender.index,
        autopct='%1.1f%%',
        colors=my_colors2,
        shadow=True,
        wedgeprops= {'edgecolor':'white', 
                     'linewidth': 3, 
                     'antialiased': True},
        textprops= {'family':'serif',
                    'color':'black',
                    'weight':'normal', 
                    'fontsize':10})
# draw circle
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
ax3 = plt.gcf()
# Adding Circle in Pie chart
ax3.gca().add_artist(centre_circle)


ax4 = fig.add_subplot(234)
ax4.set_title('Education Level', fontdict=font)
ax4.bar(education_level.index,
        education_level.values,
        color='#ef9b20',
        alpha=0.9,
        label = ['Graduate', 'High School', 'Unknown', 'Uneducated', 'College', 'Post-Graduate','Doctorate'])

plt.legend()
plt.grid(False)

ax5 = fig.add_subplot(235)
ax5.set_title('Income Category', fontdict=font)
ax5.barh(income_category.index,
         income_category.values,
         color='#ede15b',
         alpha=0.9,
         label='in thousands dollar')

plt.legend()
plt.grid(False)



plt.show()
```

[10 poin] Buatlah 1 visualisasi menggunakan countplot yang memadukan lebih dari satu kolom (bebas antar kolom yang mana saja)  

```python
sns.set(font_scale=2)
plt.rcParams["figure.figsize"] = [22,10]
plt.rcParams["figure.autolayout"] = True

sns.countplot(x='Income_Category', data= df, hue='Marital_Status', palette='pastel')
plt.show()
```

[10 poin] Buatlah 1 visualisasi menggunakan scatterplot yang memadukan lebih dari satu kolom (bebas antar kolom yang mana saja)  

```python
plt.rcParams["figure.figsize"] = [20,15]
plt.rcParams["figure.autolayout"] = True
sns.scatterplot(x='Education_Level',
                    y='Income_Category',
                    data=df,
                    size='Credit_Limit',
                    palette='pastel')
```

[10 poin] Buatlah 1 visualisasi menggunakan heatmap yang memadukan lebih dari satu kolom (bebas antar kolom yang mana saja)  

```python
df_num = df[['Customer_Age','Credit_Limit','Total_Trans_Amt']]
corr = df_num.corr()

sns.heatmap(corr, cmap='RdBu', vmin=-1, vmax=1, annot=True, annot_kws={'fontsize':20, 'fontweight':'bold'})
```
